***matrix error***

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score, max_error

def matrix_error(y_true, y_pred, n_features):
    mae = mean_absolute_error(y_true, y_pred)
    mse = mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse) 
    r2 = r2_score(y_true, y_pred)*100
    
    max_err = max_error(y_true, y_pred)    
    
    # Huber Loss
    delta = 1.0  # You can adjust delta based on your needs
    huber_loss = np.mean(np.where(np.abs(y_true - y_pred) <= delta, 
                                  0.5 * (y_true - y_pred)**2, 
                                  delta * (np.abs(y_true - y_pred) - 0.5 * delta)))
    
    # Relative Squared Error (RSE)
    rse = np.sum((y_true - y_pred) ** 2) / np.sum((y_true - np.mean(y_true)) ** 2)
    
    # Relative Absolute Error (RAE)
    rae = np.sum(np.abs(y_true - y_pred)) / np.sum(np.abs(y_true - np.mean(y_true)))

    # Adjusted R-squared
    if x_train is not None:
        n = len(y_true)
        p = x_train.shape[1]
        adj_r2 = (1 - (1 - r2 / 100) * ((n - 1) / (n - p - 1)))  # Convert Adjusted R² to percentage
    else:
        adj_r2 = 0


    return {
        'R²': r2,
        'MAE': mae,
        'MSE': mse,
        'RMSE': rmse,
        'Adj R²': adj_r2,
        'Max Error': max_err,
        'Huber Loss': huber_loss,
        'RSE': rse,
        'RAE': rae
    }
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.metrics import mean_squared_error, make_scorer
import numpy as np

# Define your models
models = {
    'Linear': LinearRegression(),
    'MLP': MLPRegressor(max_iter=500),
    'KNeighbors': KNeighborsRegressor(),
    'DecisionTree': DecisionTreeRegressor(),
    'RandomForest': RandomForestRegressor(),
    'XGBoost': XGBRegressor(),
    'AdaBoost': AdaBoostRegressor()
}

# Define 5-Fold Cross Validation and scoring
scorer = make_scorer(mean_squared_error, greater_is_better=False)  # for regression

# Store results
results = {}

for name, model in models.items():
    scores = cross_val_score(model, x_train, y_train, cv=5, scoring=scorer, n_jobs=-1)
    avg_score = np.mean(scores)
    results[name] = avg_score
    print(f"{name}: Avg CV MSE = {abs(avg_score):.4f}")

# Find best model (lowest MSE)
best_model = min(results, key=results.get)
print(f"\n✅ Best model based on 5-Fold CV: {best_model} with MSE = {abs(results[best_model]):.4f}")
```

***assemble learning***

```python
from sklearn.ensemble import BaggingRegressor, GradientBoostingRegressor, StackingRegressor, VotingRegressor

# Function to fit models, average predictions, and evaluate all error metrics
def fit_and_evaluate_combination(models_to_fit, x_train, x_test, y_train, y_test, ensemble_type='average'):
    
    if ensemble_type == 'bagging':
        # Bagging with the first model as base
        base_model = models[models_to_fit[0]]
        bagging_model = BaggingRegressor(estimator=base_model, n_estimators=10)
        bagging_model.fit(x_train, y_train)
        train_predictions = bagging_model.predict(x_train)
        test_predictions = bagging_model.predict(x_test)

    elif ensemble_type == 'boosting':
        # Boosting with a specified base model
        base_model = models[models_to_fit[0]]
        boosting_model = GradientBoostingRegressor(n_estimators=100)
        boosting_model.fit(x_train, y_train)
        train_predictions = boosting_model.predict(x_train)
        test_predictions = boosting_model.predict(x_test)

    elif ensemble_type == 'stacking':
        # Stacking Regressor with selected models
        base_models = [models[model_name] for model_name in models_to_fit]
        stacking_model = StackingRegressor(estimators=[(name, model) for name, model in zip(models_to_fit, base_models)], final_estimator=LinearRegression())
        stacking_model.fit(x_train, y_train)
        train_predictions = stacking_model.predict(x_train)
        test_predictions = stacking_model.predict(x_test)

    elif ensemble_type == 'voting':
        # Voting Regressor with selected models
        voting_model = VotingRegressor(estimators=[(name, models[name]) for name in models_to_fit])
        voting_model.fit(x_train, y_train)
        train_predictions = voting_model.predict(x_train)
        test_predictions = voting_model.predict(x_test)

    # Calculate error metrics for train and test sets
    train_metrics = matrix_error(y_train, train_predictions, x_train.shape[1])
    test_metrics = matrix_error(y_test, test_predictions, x_test.shape[1])

    return train_metrics, test_metrics
```

```python
# Loop through all combinations of 2 to 5 models and print error metrics
for num_models in range(2, 3):  
    for model_combination in combinations(models.keys(), num_models): 
        

# # Loop through all combinations of 3 models, starting from the 6th combination

# model_combinations = list(combinations(models.keys(), 3))  # Get all combinations of 3 models
# for i, model_combination in enumerate(model_combinations):
#     if i < 5:  # Skip the first 6 combinations
#         continue

        
    # Bagging Ensemble
        train_metrics, test_metrics = fit_and_evaluate_combination(model_combination, x_train, x_test, y_train, y_test, ensemble_type='bagging')
        print(f"Bagging Combination: {model_combination} - Train Metrics: {train_metrics}")
        print(f"Bagging Combination: {model_combination} - Test Metrics: {test_metrics}\n")

        # Boosting Ensemble
        train_metrics, test_metrics = fit_and_evaluate_combination(model_combination, x_train, x_test, y_train, y_test, ensemble_type='boosting')
        print(f"Boosting Combination: {model_combination} - Train Metrics: {train_metrics}")
        print(f"Boosting Combination: {model_combination} - Test Metrics: {test_metrics}\n")

        # Stacking Ensemble
        train_metrics, test_metrics = fit_and_evaluate_combination(model_combination, x_train, x_test, y_train, y_test, ensemble_type='stacking')
        print(f"Stacking Combination: {model_combination} - Train Metrics: {train_metrics}")
        print(f"Stacking Combination: {model_combination} - Test Metrics: {test_metrics}\n")

        # Voting Ensemble
        train_metrics, test_metrics = fit_and_evaluate_combination(model_combination, x_train, x_test, y_train, y_test, ensemble_type='voting')
        print(f"Voting Combination: {model_combination} - Train Metrics: {train_metrics}")
        print(f"Voting Combination: {model_combination} - Test Metrics: {test_metrics}\n")

        print("-" * 40)

        results = []
        results.append({
            "Combination": model_combination,
            f"{model_combination} Train": train_metrics,
            f"{model_combination} Test": test_metrics,
        })
        results_df = pd.DataFrame(results)
        results_df.to_csv("ensemble_results.csv", index=False)
```
